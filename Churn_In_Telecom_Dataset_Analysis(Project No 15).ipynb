```python
# Dropping some columns which are less important
df.drop(['area code','phone number'],axis=1,inplace=True)
```

```python
# Using LabelEncoder to convert categorical data into numeric
from sklearn.preprocessing import LabelEncoder
cols=['state','international plan',"voice mail plan",'churn']
le=LabelEncoder()
for col in cols:
    df[col]=le.fit_transform(df[col])
```

```python
df.head()
```

```python
# Checking And Removing Outliers
print("data shape before zscore", df.shape)
from scipy.stats import zscore
z=np.abs(zscore(df))
df=df[(z<3).all(axis=1)]
print("data shape after zscore", df.shape)
```

```python
# Splitting Data into x and y
x=df.iloc[:,0:-1]
y=pd.DataFrame(df.iloc[:,-1])
print(x.shape)
print(y.shape)
```

```python
# cheking Skewness
x.skew()
```

```python
# for removing skewness use log method
for i in x.columns:
    if x.skew()[i]>0.55:
        x[i]=np.log1p(x[i])
```

```python
x.skew()
```

```python
#as the skewness needs to be reduced again, we use sqrt function to reduce
for i in x.columns:
    if x.skew()[i]>0.55:
        x[i]=np.sqrt(x[i])
```

```python
x.skew()
```

```python
# as skewness is present so I can use standard scalling
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x=scaler.fit_transform(x)
x=pd.DataFrame(x)
x.head()
```

# Train-Test Split


```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
print(x_train.shape,x_test.shape)
print(y_train.shape,y_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import cross_val_score
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score,roc_curve,auc
```

```python
from sklearn.model_selection import GridSearchCV
KNN=KNeighborsClassifier()
parameters={'n_neighbors':(1,10)}
gdk=GridSearchCV(KNN,parameters)
gdk.fit(x_train,y_train)
gdk.best_params_
```

```python
#Checking gridsearch for different algorithms
SV=SVC()
parameters={'kernel':('linear','rbf'), 'C':[1,10]}
gdsr=GridSearchCV(SV,parameters)
gdsr.fit(x_train,y_train)
gdsr.best_params_
```

```python
KNN=KNeighborsClassifier(n_neighbors=10)
SV=SVC(kernel='rbf',C=10)
LR=LogisticRegression()
DT=DecisionTreeClassifier(random_state=6)
GNB=GaussianNB()
```

```python
models=[]
models.append(("KNeighborsClassifier",KNN))
models.append(("SVC",SV))
models.append(("LogisticRegression",LR))
models.append(("DecisionTreeClassifier",DT))
models.append(("GaussianNB",GNB))
```

```python
Model=[]
score=[]
cvs=[]
rocscore=[]
for name,model in models:
    print("-----------------------",name,"----------------------")
    print("\n")
    Model.append(name)
    model.fit(x_train,y_train)
    print(model)
    pred=model.predict(x_test)
    print("\n")
    AS=accuracy_score(y_test,pred)
    print("Accuracy_Score =",AS)
    score.append(AS*100)
    print("\n")
    sc=cross_val_score(model,x,y,cv=5,scoring="accuracy").mean()
    print("Cross_Val_Score =",sc)
    cvs.append(sc*100)
    print("\n")
    false_positive_rate,true_positive_rate,thresholds=roc_curve(y_test,pred)
    roc_auc=auc(false_positive_rate,true_positive_rate)
    print("roc_auc_score =",roc_auc)
    rocscore.append(roc_auc*100)
    print("\n")
    print("classification_report\n",classification_report(y_test,pred))
    print("\n")
    cm=confusion_matrix(y_test,pred)
    print(cm)
    print("\n")
    plt.figure(figsize=(10,40))
    plt.subplot(911)
    plt.title(name)
    print(sns.heatmap(cm,annot=True))
    plt.subplot(912)
    plt.title(name)
    plt.plot(false_positive_rate,true_positive_rate,label="AUC=%0.2f"%roc_auc)
    plt.plot([0,1],[0,1],"r--")
    plt.legend(loc="lower right")
    plt.ylabel("True Positive Rate")
    plt.xlabel("False Positive Rate")
    print("\n\n")
```

```python
result=pd.DataFrame({"Model":Model,"Accuracy_Score":score,"cross_val_score":cvs,"Roc_Auc_Curve":rocscore})
result
```

```python
# As we see in above models "SVC" is working well till now. 
# Lets use ensemble techniques
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
rf.fit(x_train,y_train)
predrf=rf.predict(x_test)
print(accuracy_score(y_test,predrf))
print(confusion_matrix(y_test,predrf))
print(classification_report(y_test,predrf))
```

```python
score_rf=cross_val_score(rf,x,y,cv=10,scoring="accuracy")
print("Accuracy_score",score_rf)
print("Mean_scores",score_rf.mean()*100)
print("Standard_Deviation",score_rf.std())
```

```python
from sklearn.ensemble import AdaBoostClassifier
ad=AdaBoostClassifier(n_estimators=70)
ad.fit(x_train,y_train)
predad=ad.predict(x_test)
print(accuracy_score(y_test,predad))
print(confusion_matrix(y_test,predad))
print(classification_report(y_test,predad))
```

```python
score_ad=cross_val_score(ad,x,y,cv=10,scoring="accuracy")
print("Accuracy_score",score_ad)
print("Mean_scores",score_ad.mean()*100)
print("Standard_Deviation",score_ad.std())
```

after applying so many model I found that model "RandomForestClassifier()" have maximum Accuracy Score, Cross_val_score. so I choose "RandomForestClassifier" (rf) because it performs very well on this dataset.
